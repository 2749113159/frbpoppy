```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colormaps
from matplotlib.colors import ListedColormap
import cmaps

import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.loadtxt('mc_3_11_1000.txt', unpack=True, usecols=(0,1,2,3,4,5)).T
df = pd.DataFrame(data, columns=['alpha', 'si', 'li', 'n_frb', 'n_frb_ks', 'p_val'])
df = df.fillna(0)
```

```python
NUM = 11
r = range(NUM)
alphas = np.linspace(-3.0, -1.0, NUM)
sis = np.linspace(-2.0, 2.0, NUM)
lis = np.linspace(-2.0, -0.4, NUM)
```

```python
U = np.zeros([11, 11, 11])
for i in range(11):
    for j in range(11):
        for k in range(11):
            U[i][j][k] = df[df.alpha-alphas[10-i]<1e-3][df.si-sis[10-j]<1e-3][df.li-lis[10-k]<1e-3].mean().p_val
```

```python
X, Y, Z = np.meshgrid(alphas, sis, lis)
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
img = ax.scatter3D(X, Y, Z, c=U, alpha=0.66, cmap=cmaps.NCV_jet, marker='.', s=U*5000)
#img = ax.scatter3D(X, Y, Z, c=U, alpha=0.8, cmap='RdBu_r', marker='.', s=U*5000)
ax.set_xlabel('alpha')
ax.set_ylabel('si')
ax.set_zlabel('li')
fig.colorbar(img)
fig.savefig('mc_alpha.pdf', bbox_inches='tight', transparent=True)
plt.show()
```
