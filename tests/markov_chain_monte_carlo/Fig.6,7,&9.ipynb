```python
import numpy as np
import corner
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

# Fig.6

```python
flatchainpl_1 = np.loadtxt('sampler/mcmc_sampler_vol_co_20000_1.txt')
flatchainpl_2 = np.loadtxt('sampler/mcmc_sampler_vol_co_20000_2.txt')
flatchainpl_3 = np.loadtxt('sampler/mcmc_sampler_vol_co_20000_3.txt')
flatchainpl_4 = np.loadtxt('sampler/mcmc_sampler_vol_co_20000_4.txt')
flatchainpl_5 = np.loadtxt('sampler/mcmc_sampler_vol_co_20000_5.txt')
flatchainpl_6 = np.loadtxt('sampler/mcmc_sampler_vol_co_20000_6.txt')
ndiscard = 2000 ## Discard 10% initial points
flatchainpl = np.vstack((flatchainpl_1[ndiscard:, :], flatchainpl_2[ndiscard:, :], flatchainpl_3[ndiscard:, :],\
                         flatchainpl_4[ndiscard:, :], flatchainpl_5[ndiscard:, :], flatchainpl_6[ndiscard:, :]))
```

```python
fig = corner.corner(flatchainpl,\
                    labels=['$\\alpha$', '$li$', '$\mathrm{log}_{10} (w_\mathrm{mean}/\mathrm{ms})$', '$w_\mathrm{std}$ (ms)', '$\mathrm{DM}_\mathrm{igm, slope}$\n (pc cm$^{-3}$)', '$\mathrm{DM}_\mathrm{host, mean}$\n (pc cm$^{-3}$)', '$\mathrm{DM}_\mathrm{host, std}$\n (pc cm$^{-3}$)'],\
                    dpi_plot=400, color='#2B6688', smooth=2.0,\
                    truths=[-1.56, -1.53, -0.45, 1.75, 800, 570, 580], truth_color='#76A2BB', \
                    label_kwargs=dict(fontsize='x-large'), show_titles=False)
fig.savefig('mcmc_pl.pdf', transparent=True, bbox_inches='tight')
```
